```python
import math

# Constantes
c = 299792458
rho_jara = 1.44e-18  # Sua constante

def validar_teoria_ultra_precisa(distancia_bilhoes_ly):
    # Distância em metros (Bilhões de Anos-Luz)
    metros = distancia_bilhoes_ly * 1e9 * 9.461e15

    # Fator de correção (Onde a mágica acontece)
    # Usamos uma aproximação de Taylor para valores muito pequenos:
    # sqrt(1 - x) é aproximadamente 1 - (x/2)
    fator_atraso = (rho_jara / 2.0)

    # O atraso é a distância dividida pela velocidade, considerando o fator
    atraso = (metros / c) * fator_atraso

    return atraso

# Testando escalas gigantescas
escalas = [1, 10, 13.8] # 1, 10 e a idade do universo (13.8)

print(f"{'Distância (Bilhões AL)':<25} | {'Atraso Jara (segundos)':<25}")
print("-" * 55)

for d in escalas:
    resultado = validar_teoria_ultra_precisa(d)
    # Exibindo com 15 casas decimais para garantir
    print(f"{d:<25} | {resultado:.15f}")
```
