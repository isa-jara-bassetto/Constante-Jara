```python
import matplotlib.pyplot as plt
import numpy as np

# Constantes do seu modelo
c = 299792458
rho_jara = 1.44e-18

# Criando os dados
distancia = np.linspace(0, 10, 100) # Bilhes de anos-luz
v_einstein = np.full(100, c)
v_jara = c * (1 - (rho_jara * distancia * 1e7)) # Decaimento baseado na sua constante

# Criando o grfico
plt.figure(figsize=(10, 5))
plt.plot(distancia, v_einstein, 'r--', label='Einstein (Constante)', linewidth=2)
plt.plot(distancia, v_jara, 'b-', label='Modelo Jara (Varivel)', linewidth=2)

plt.title('Simulao: Impacto da Constante Jara na Velocidade da Luz', fontsize=14)
plt.xlabel('Distncia Cosmolgica (Bilhes de Anos-Luz)', fontsize=12)
plt.ylabel('Velocidade (m/s)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# Salvar a imagem
plt.savefig('grafico_jara.png', dpi=300)
plt.show()
```
